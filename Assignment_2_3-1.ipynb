3. Uncovering factors that lead to employee attrition

```python
# (a) breakdown of distance from home by job role and attrition using Groupby() function.
```

```python
# Select the columns "Attrition", "JobRole", and "DistanceFromHome" from the data frame
df = df[["Attrition", "JobRole", "DistanceFromHome"]]
```

```python
df
```

```python
# Group the data frame by "Attrition" and "JobRole, aggregated by distance using the describe function
dist_from_home_by_jrole_attrition = df.groupby(["JobRole", "Attrition"])["DistanceFromHome"].describe()
```

```python
dist_from_home_by_jrole_attrition
```

```python
# Instead of the predefined statistics, you can use
# specific combinations of aggregating statistics for given columns can be defined using the DataFrame.agg() method:
# e.g df.agg({'age': ['min', 'max', 'median', 'skew'],
        
df.agg({'JobRole': ['min', 'max', 'median', 'skew'],
             'DistanceFromHome': ['min', 'max', 'median', 'mean'],
            'Attrition': ['min', 'max', 'median', 'mean']})
```

```python
# Group the data frame by "Attrition" and "JobRole, aggregated by mean distance"
dist_from_home_by_jrole_attrition = df.groupby(["JobRole", "Attrition"])["DistanceFromHome"].mean()
```

```python
dist_from_home_by_jrole_attrition
```

```python
# Group the data frame by "Attrition" and "JobRole, aggregated by median distance"
dist_from_home_by_jrole_attrition = df.groupby(["JobRole", "Attrition"])["DistanceFromHome"].median()
```

```python
dist_from_home_by_jrole_attrition
```

```python
# Group the data frame by "Attrition" and "JobRole, aggregated by minimum distance"
dist_from_home_by_jrole_attrition = df.groupby(["JobRole", "Attrition"])["DistanceFromHome"].min()
```

```python
dist_from_home_by_jrole_attrition
```

```python
# Group the data frame by "Attrition" and "JobRole, aggregated by maximum distance"
dist_from_home_by_jrole_attrition = df.groupby(["JobRole", "Attrition"])["DistanceFromHome"].max()
```

```python
dist_from_home_by_jrole_attrition
```

```python
# Group the data frame by "Attrition" and "JobRole, aggregated by the variance of distance"
dist_from_home_by_jrole_attrition = df.groupby(["JobRole", "Attrition"])["DistanceFromHome"].var()
```

```python
dist_from_home_by_jrole_attrition
```

```python
# Group the data frame by "Attrition" and "JobRole, aggregated by the standard deviation of distance"
dist_from_home_by_jrole_attrition = df.groupby(["JobRole", "Attrition"])["DistanceFromHome"].std()
```

```python
dist_from_home_by_jrole_attrition
```

```python
# (b) breakdown of average monthly income by education and attrition using Groupby()
```

```python
# Select the columns "Attrition", "Education", and "MonthlyIncome" from the data frame
df=pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

df = df[["Education", "Attrition", "MonthlyIncome"]]
```

```python
df
```

```python
# Group the data frame by "Attrition" and "Education", aggregated by "MonthlyIncome" using the describe function
mon_inc_by_educ_attrition = df.groupby(["Education", "Attrition"])["MonthlyIncome"].describe()
```

```python
mon_inc_by_educ_attrition
```

```python
# Group the data frame by "Attrition" and "Education, aggregated by mean monthly Income"
mon_inc_by_educ_attrition = df.groupby(["Education", "Attrition"])["MonthlyIncome"].mean()
```
